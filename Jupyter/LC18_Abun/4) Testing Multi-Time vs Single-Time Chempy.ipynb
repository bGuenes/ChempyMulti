```python
%pylab inline
import os
os.chdir('/home/oliverphilcox/ChempyMulti/')
from Chempy.parameter import ModelParameters
a=ModelParameters()
from Chempy.cem_function import multi_timestep_chempy,single_timestep_chempy,cem_real2_all_times,cem_real2_single_time
```
